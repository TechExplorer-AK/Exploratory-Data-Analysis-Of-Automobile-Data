## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a copy of the current dataset and assigning to df
df=dataset.copy()

# Replace all column values '?' with NaN.
df=df.replace('?',np.NaN)
````

````python
df.dtypes
````

````python
# After replacing '?' with NaN, there are some null values exist in new dataset 'df'.
# Missing/ Null values count
df.isnull().sum()
````

````python
# Visualizing the missing/null values after replacing '?' with NaN
sns.heatmap(df.isnull(),cbar=False)
````

````python
# convert column variables 'object' datatype to 'float64' datatype
df['price']=df['price'].astype('float64')
df['horsepower']=df['horsepower'].astype('float64')
df['normalized-losses']=df['normalized-losses'].astype('float64')
````

````python
# # Identify outliers in each numeric column variable using box plot
plt.figure(figsize=(14, 10))
numerical_columns = df.select_dtypes(include=[np.number]).columns

# Iterate over numerical columns and create boxplots
for i, column in enumerate(numerical_columns, 1):
    ax = plt.subplot(5, 5, i)
    sns.boxplot(y=df[column], orient='v', width=0.5, ax=ax)
    ax.set_ylabel(column, fontsize=12)  # Set ylabel on the Axes object (ax)

plt.tight_layout()
plt.show()
````

````python
df.dtypes
````

````python
# Drop missing/null values from dataset df
df= df.dropna()
````

````python
# check columns datatype
df.dtypes
````

````python
# # Convert column variables 'float64' datatype to 'int64' datatype

df['horsepower']=df['horsepower'].astype('int64')
df['normalized-losses']=df['normalized-losses'].astype('int64')
````

````python
# check whether datatypes changed for above column variables or not
df.dtypes
````

````python
# Check maximum price for each car brand
max_price=df.groupby(['make'])['price'].max()
print(max_price)
````

````python
# check average price for each car brand
average_price=df.groupby(['make'])['price'].mean().round(2)
print(average_price)
````

````python
# calulate average price for each unique car body style
car_body_style_avg_price= df.groupby(['body-style'])['price'].mean().round(2).sort_values()
print(car_body_style_avg_price)
````

````python
# convert 'car_body_style_avg_price' in a dataframe
car_body_style_avg_price_df= pd.DataFrame(car_body_style_avg_price).reset_index()
car_body_style_avg_price_df.columns=['body-style','avg_price']
print(car_body_style_avg_price_df)
````

````python
# Calculate correlation coefficient
# Correlation between City MPG and Price
correlation_city = df['city-mpg'].corr(df['price'])
print(f"Correlation between City MPG and Price: {correlation_city}")
````

````python
# Calculate correlation coefficient
# Correlation between Highway MPG and Price
correlation_highway = df['highway-mpg'].corr(df['price'])
print(f"Correlation between Highway MPG and Price: {correlation_highway}")
````

#### Chart - 2- Relationship between normalized losses and car prices.

````python
# Chart - 2 visualization code
# Scatter plot of normalized losses vs price
# Create figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Scatter plot
plt.scatter(df['normalized-losses'], df['price'], s=30, marker='o', color='green', alpha=0.7)

# Set title, x-axis label, and y-axis label using ax methods
ax.set_title('Normalized Losses vs. Price', fontweight='bold', color='olive')
ax.set_xlabel('Normalized Losses', fontweight='bold', color='olive')
ax.set_ylabel('Price', fontweight='bold', color='olive')

# Customize x-axis ticks
ax.set_xticks(np.arange(0, 257, 30))
plt.grid(True)
````

##### 1. Why did you pick the specific chart?

A scatter plot is a type of graph that displays the relationship between two numerical variables.

And, here we have also two numerical column variables to visualize the relationship between these two.


##### 2. What is/are the insight(s) found from the chart?

Here, it shows the relationship between normalized losses and car prices.

**Normalized losses** defined as Average loss payment(amount paid for insurance claims against baseline) per insured vehicle per year.

**It has a standard value(baseline value) of 100.**

**Values Above 100**: If a vehicle's normalized-losses are above 100, it indicates that the average loss payment for that vehicle model is higher than the baseline average. It means vehicle may be associated with higher insurance costs due to factors like repair costs, accident frequency, or theft rates.

**Values Below 100**: if normalized-losses are below 100, it means vehicle has lower average loss than baseline average.It means vehicle may be associated with lower insurance costs due to factors such as  lower repair costs, or lower accident rates .

**Let's take an example**:-

Calculation for Honda brand:
   

*  Normalized-losses = 110
*  Difference from baseline = 110 - 100 = 10

* Percentage difference = (10 / 100) * 100% = 10%

  So, the Honda has normalized-losses 10% above the baseline.

 Companies can adjust prices based on whether a vehicle's normalized losses are above or below 100. Vehicles with lower normalized losses (**below 100**) can be priced higher, reflecting lower expected insurance costs and potentially increasing profit margins.

 Conversely, vehicles with higher normalized losses (**above 100**) might be priced competitively to account for higher insurance risks, attracting customers looking for more comprehensive insurance coverage.
